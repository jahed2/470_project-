```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as scl
hb=1.0546*10**-34
m=9.11*10**-31
N = 400
L = 8*10**-9
b = 14*10**-9
x = np.linspace(-b,b,N)
h = x[1]-x[0] 
V0 = 8.812*10**-19
V=np.zeros(N)
for i in range(N):
    if x[i]> -L and x[i]< L:
        V[i]= V0

M = 1./(h*h)*(np.diag(np.ones(N-1),-1) -2* np.diag(np.ones(N),0) + np.diag(np.ones(N-1),1))
H = -(hb*hb)/(2.0*m)*M + np.diag(V) 
Energy,psiT = np.linalg.eigh(H)
psi = np.transpose(psiT)
#

print(Energy)
print(np.min(Energy))
print(np.max(Energy))
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=False)
plt.rc('axes', axisbelow=True)
plt.figure(figsize=(12,7))
plt.xlim(-4*b,4*b)
for i in range(5):
    if E[i]>0:
        if psi[i][N-9]<0:
            plt.plot(x,-psi[i]/np.sqrt(h),label="$E_{}$={:>10.3f}".format(i,E[i]))
        else:
            plt.plot(x,psi[i]/np.sqrt(h),label="$E_{}$={:>10.3f}".format(i,E[i]))
plt.title("solutions to the finite square well")
plt.legend()
plt.savefig("finite_square_well_wavefunctions.pdf")
plt.show()
```
